```python
# !pip3 install librosa
import librosa

y, sr = librosa.load("./smeshariki-kardebalet.mp3")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(y[:])
```

```python
plt.plot(y[536000:537000])
```

```python
step=sr*25.0/1000
```

```python
step=int(step)
```

```python
spectrogram=librosa.feature.melspectrogram(y,sr,n_mels=64,hop_length=step)
```

```python
import numpy as np
```

```python
spectrogram[:,2500:3000].shape
```

```python
plt.imshow((spectrogram[:,2500:3000]))
plt.colorbar()
```

```python
plt.figure(figsize=(24,8))
start = 2500
plt.imshow(np.log(spectrogram[:,start:start+500]))
```
